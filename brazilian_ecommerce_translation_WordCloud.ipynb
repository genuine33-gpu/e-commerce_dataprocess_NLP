```python
def translate_text(text):
    if pd.isnull(text) or text.lower() == "nan":
        return text
    else:
        translator = Translator()
        try:
            translated = translator.translate(text, src='pt', dest='en')
            return translated.text
        except Exception as e:
            return None

```

```python
review_data = pd.DataFrame(review_data, columns=['review_comment_message'])
```

```python
import nltk
nltk.download('punkt')
```

```python


# Veriyi y√ºkleyin veya var olan veriyi kullanƒ±n
# ... (Veri y√ºkleme kodu)

comments = []
stop_words = set(stopwords.words('portuguese'))

for words in review_data['review_comment_message']:
    only_letters = re.sub("[^a-zA-Z]", " ", words)
    tokens = nltk.word_tokenize(only_letters) # tokenize the sentences
    lower_case = [l.lower() for l in tokens] # convert all letters to lower case
    filtered_result = list(filter(lambda l: l not in stop_words, lower_case)) # Remove stopwords from the comments
    comments.append(' '.join(filtered_result))
```

```python
#Using wordcloud to visualize the comments
unique_string=(" ").join(comments)
wordcloud = WordCloud(width = 2000, height = 1000,background_color='white').generate(unique_string)
plt.figure(figsize=(20,12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
co = CountVectorizer(ngram_range=(3,3))
counts = co.fit_transform(comments)
important_trigrams = pd.DataFrame(counts.sum(axis=0),columns=co.get_feature_names_out()).T.sort_values(0,ascending=False).head(50)

important_trigrams = important_trigrams.reset_index()
important_trigrams.rename(columns={'index':'trigrams',0:'frequency'}, inplace=True)
important_trigrams
```

```python
translated_trigrams = important_trigrams['trigrams'].apply(translate_text)
important_trigrams['english_translation'] = translated_trigrams
important_trigrams
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# WordCloud nesnesini olu≈üturun
wordcloud = WordCloud(width=2000, height=1000, background_color='white')

# G√∂rselle≈ütirmeyi yapƒ±n
wordcloud_image = wordcloud.generate_from_frequencies({row['english_translation']: row['frequency'] for index, row in important_trigrams.iterrows()})

# WordCloud'ƒ± g√∂r√ºnt√ºleyin
plt.figure(figsize=(20, 12))
plt.imshow(wordcloud_image, interpolation='bilinear')
plt.axis("off")
plt.show()
```
